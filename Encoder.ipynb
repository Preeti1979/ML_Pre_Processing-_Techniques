```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
data = pd.read_excel('Employee_Data.xls')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.Department.value_counts()
```

```python
data.Department = data.Department.fillna('Sales and Marketing')
```

```python
data.Department.value_counts()
```

```python
## Label Encoder
```

```python
data.Department = data.Department.astype('category')  # Changing the type of Department to category
data.Department = data.Department.cat.codes           # Assigning codes 0,1,2 to department in alphabetical order of category
```

```python
data.Department.value_counts()
```

```python
# One Hot encoder
data = pd.get_dummies(data, columns = ['Department'])
```
